```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from nltk import tokenize
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt

def generate_wdc(text):
    wc = WordCloud(font_path='framd.ttf',
                  width = 2400, height = 1800, ranks_only=None, relative_scaling=0.5,
                  stopwords=set(STOPWORDS)).generate(text)
    
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
```

```python
f=open("obama.txt","r")
line=f.read()
generate_wdc(line)
```
